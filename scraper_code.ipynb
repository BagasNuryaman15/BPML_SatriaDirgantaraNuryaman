# **Import Library**

```python
from google_play_scraper import app, reviews_all, Sort 
import pandas as pd
```

# **Scraping Code**

```python
# Mengambil data aplikasi freefire dari Google Play Store
scraping_views = reviews_all(
    'com.dts.freefireth', # FreeFire Id
    lang = 'id', # Mengambil review dalam bahasa Indonesia
    country = 'id', # Mengambil review dari negara Indonesia
    sort = Sort.NEWEST, # Mengambil review terbaru
    count = 2000, # Mengambil 1000 review
)

# Membatasi jumlah review yang diambil
if len(scraping_views) > 10000:
    scraping_views = scraping_views[:10000]

print(f'Pengumpulan review done bang! Total review: {len(scraping_views)}')
```

# **Loading Data Hasil Scraping**

```python
ff_df = pd.DataFrame(scraping_views)

# Menampilkan hasil scraping
display(ff_df.head(), ff_df.tail())
```

```python
ff_df.to_csv('ff_scraping.csv', index=False)
```
